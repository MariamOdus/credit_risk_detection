### Imports ###

```python
#Data Manipulation Libraries
import numpy as np
import pandas as pd

#Data Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
import plotly.express as px
import plotly.graph_objects as go

#Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

### Data Visualisations

```python
#Creating data
data = {
    'Year': ['A', 'B', 'C', 'D', 'E'],
    'Unemployment Rate': [10, 15, 7, 12, 20]
}
#Creating DataFrame
df = pd.DataFrame(data)
```

#### Line Charts 

```python
from matplotlib import lines

plt.plot(df['Year'], df['Unemployment Rate'],
          marker='o', color='blue',
          linewidth = 0.75, linestyle='--',
          markersize=5, markerfacecolor='pink', markeredgecolor='green',
          markeredgewidth=1.5)

plt.title('Unemployment Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.grid()
plt.show()
```

#### Bar Chart

```python
dataA = [1,5,3,10,5,15,7,20,9,25]
dataB = [2,6,4,11,6,16,8,21,10,26]

#Bar Chart - Vertical
plt.bar(range(len(dataA)), dataA, color='lightgreen',edgecolor='darkgreen', alpha=0.5, label='Dataset A')
plt.title('Bar Chart Example')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()

#Bar Chart - Horizontal
plt.barh(range(len(dataB)), dataA, color='beige', alpha=0.5, edgecolor='black', label='Dataset B')
plt.title('Bar Chart Example')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()

#Stacked Bar Chart
plt.bar(range(len(dataA)), dataA, color='lightgreen',edgecolor='darkgreen', alpha=0.5, label='Dataset A')
plt.bar(range(len(dataB)), dataB, color='beige', edgecolor='black', alpha=0.5, label='Dataset B', bottom=dataA)
plt.title('Bar Chart Example')
plt.xlabel('Index')
plt.ylabel('Value')
plt.show()
```

#### Histograms

```python
data1 = np.random.rand(100)
data2 = np.random.rand(100)


#Histogram
plt.hist(data1, bins=15, color='green', edgecolor='darkgreen', alpha=0.2, label='Dataset A', histtype='bar', orientation='vertical', density=True)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

#Stacked Histogram
plt.hist([data1, data2], bins=15, color=['green', 'pink'], edgecolor=['darkgreen', 'purple'], alpha=0.2, label=['Dataset A', 'Dataset B'], histtype='bar', orientation='vertical', density=True, stacked=True)
plt.title('Stacked Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

#### Pie Chart

```python
pieData = [30, 20, 15, 10, 25]
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']

plt.figure(figsize=(10, 6))
plt.pie(pieData, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'pink', 'lightblue', 'orange', 'beige'], explode=(0, 0, 0, 0., 0.1), shadow=True)
plt.title('Pie Chart Example')
plt.axis('equal')
plt.legend(loc='upper right')
plt.show()
```

#### Area Chart 

```python
AreaData1 = {
    'Year': ['A', 'B', 'C', 'D', 'E'],
    'Unemployment Rate': [10, 15, 7, 12, 20]
}

AreaData2 = {
    'Year': ['A', 'B', 'C', 'D', 'E'],
    'Unemployment Rate': [8, 12, 5, 10, 18]
}

df = pd.DataFrame(AreaData1) 

plt.fill_between(df['Year'], df['Unemployment Rate'], color='lightblue', alpha=0.5)
plt.plot(df['Year'], df['Unemployment Rate'], marker='o', color='blue', linewidth=1.5, label='Unemployment Rate')
plt.title('Area Chart Example')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.grid()
plt.legend()
plt.show()
```

#### Scatter Plot 

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (20 * np.random.rand(N))**2

plt.scatter(x, y, s=area, 
            c = np.random.rand(N),edgecolor='darkgreen',
            marker='o', alpha=0.3, 
            label='Data Points')

plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.legend()
plt.show()

#clustered scatter plot
N = 50
x1 = np.random.rand(N)
y1 = np.random.rand(N)
x2 = np.random.rand(N)
y2 = np.random.rand(N)
plt.scatter(x1, y1, s=area, color='green', edgecolor='darkgreen', marker='o', alpha=0.3, label='Dataset A')
plt.scatter(x2, y2, s=area, color='pink', edgecolor='darkred', marker='o', alpha=0.3, label='Dataset B')
plt.title('Clustered Scatter Plot')
plt.tick_params(axis='both', which='major', labelsize=10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.legend()
plt.show()
```

#### Heatmap

```python
#Random Data for Correlation Matrix
corr = np.corrcoef(np.random.randn(10,100))

#Mask to hide upper triangle
mask = np.zeros_like(corr) #Create a mask of zeros with the same shape as the correlation matrix
mask[np.triu_indices_from(mask)] = True #Set the upper triangle of the mask to True, which will hide those values in the heatmap

with sns.axes_style("white"):
    ax = sns.heatmap(corr, mask=mask, cmap='PiYG', annot=True, fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

#### Kernal Density Estimate 

```python
data = sns.load_dataset('tips')
data.tail(5)
```

```python

sns.kdeplot(data, x = 'total_bill', 
            hue='time', 
            multiple='stack', 
            palette='Set2',
            alpha =0.3, 
            fill=True) 
plt.title('Stacked KDE Plot')
plt.xlabel('Total Bill')
plt.ylabel('Density')
plt.show()
```

```python
sns.kdeplot(data, x = 'total_bill', 
            hue='size', 
            multiple='fill', 
            palette='Set2',
            alpha=0.3, 
            edgecolor='grey') 
```

```python
sns.kdeplot(data, x = 'total_bill', 
            hue='size', 
            fill='true', 
            alpha=0.3,
            palette='Set2') 
```
